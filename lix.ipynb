```python
import spacy

def calculate_lix(text):
    nlp = spacy.load("fi_core_news_md")
    doc = nlp(text)
    
    # Count sentences, words, and long words
    num_sentences = len(list(doc.sents))
    num_words = len([token for token in doc if token.is_alpha])  # Count only alphabetic tokens
    num_long_words = len([token for token in doc if token.is_alpha and len(token) >= 7])
    
    # Avoid division by zero
    if num_sentences == 0 or num_words == 0:
        return None
    
    # Compute LIX
    lix = (num_words / num_sentences) + (num_long_words / num_words) * 100
    return lix

# Example text
text = 'Ilma liikkuu ympärillämme koko ajan lämpötilaerojen vaikutuksesta. Koska ilmaa ei voi nähdä, tämän ilmiön havainnoiminen voi tuntua hankalalta. Vaikka ilman liikkeen havainnointi on haastavaa, voimme havaita sen vaikutuksia esimerkiksi tuulena. Tuuli, jonka tunnemme suoraan kasvoillamme, on pääosin ilman vaakasuuntaista liikettä. Ilma liikkuu kuitenkin myös pystysuunnassa, mikä on tärkeää ymmärtää tuulen syntymisen kannalta.'
lix_score = calculate_lix(text)
print(f"LIX Score: {lix_score}")
```
